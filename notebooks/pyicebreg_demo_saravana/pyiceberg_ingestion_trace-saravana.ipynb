```python
def setup_iceberg_table():
    s3_bucket_name = "messageingest"
    catalog = load_catalog(
        "hive",
        **{ "type":"hive",
            "s3.region":"ind-south-2",
            "uri": "thrift://hive-ms-trino.apps.zagaopenshift.zagaopensource.com:9888",             #thrift://192.168.31.102:9083"
            "s3.endpoint": "http://minio-lb.apps.zagaopenshift.zagaopensource.com:9009",#"http://minio-api.apps.zagaopenshift.zagaopensource.com",
            "py-io-impl": "pyiceberg.io.pyarrow.PyArrowFileIO",
            "s3.access-key-id": "minio",
            "s3.secret-access-key": "minio123$",
            "profile": "my_profile"
            
        }
    )

    catalog.create_namespace("new")
    print(catalog.list_namespaces())

    schema = Schema(
            NestedField(field_id=1, name="traceId", field_type=StringType(), required=True),
            NestedField(field_id=2, name="createdTime", field_type=TimestampType(), required=True),
            NestedField(field_id=3, name="traceData", field_type=StringType(), required=True),
            NestedField(field_id=4, name="serviceName", field_type=StringType(), required=False),
            NestedField(field_id=5, name="startUnixNanoTime", field_type=LongType(), required=False),
            NestedField(field_id=6, name="endUnixNanoTime", field_type=LongType(), required=False),
            NestedField(field_id=7, name="duration", field_type=IntegerType(), required=False),
            NestedField(field_id=8, name="requestRoute", field_type=StringType(), required=False),
            NestedField(field_id=9, name="requestMethod", field_type=StringType(), required=False),
            NestedField(field_id=10, name="responseStatusCode", field_type=IntegerType(), required=False),
            NestedField(field_id=11, name="language", field_type=StringType(), required=False),
        )
    
    partition_spec = PartitionSpec(
        PartitionField(
            source_id=2, field_id=1000, transform=DayTransform(), name="createdTime_day"
        )
    )
    sort_order = SortOrder(SortField(source_id=1, transform=IdentityTransform()))

    table_identifier = "new.newdata"

    try:
        # Check if the table already exists
        existing_table = catalog.load_table(table_identifier)
        if existing_table:
            logger.info(f"Table {table_identifier} already exists.")
            print(f"Table {table_identifier} already exists.")
            return existing_table
    except Exception as e:
        logger.info(f"Table {table_identifier} does not exist. Creating new table.")

    # Create the Iceberg table if it does not exist
    table = catalog.create_table(
        identifier=table_identifier,
        schema=schema,
        location=f"s3a://messageingest/newdata",
        partition_spec=partition_spec,
        sort_order=sort_order
    )
    
    return table
```

```python
import pyarrow as pa
from pyiceberg.io.pyarrow import PyArrowFileIO
def write_to_iceberg(table, messages):

    # Convert messages to PyArrow Table?
    data = {
        "traceId": [msg.get("traceId") for msg in messages],
        "createdTime": [msg.get("createdTime") for msg in messages],
        "traceData": [json.dumps(msg) for msg in messages],  # Store the entire message as a string
        "serviceName": [msg.get("serviceName") for msg in messages],
        "startUnixNanoTime": [msg.get("startUnixNanoTime") for msg in messages],
        "endUnixNanoTime": [msg.get("endUnixNanoTime") for msg in messages],
        "duration": [msg.get("duration") for msg in messages],
        "requestRoute": [msg.get("requestRoute") for msg in messages],
        "requestMethod": [msg.get("requestMethod") for msg in messages],
        "responseStatusCode": [msg.get("responseStatusCode") for msg in messages],
        "language": [msg.get("language") for msg in messages],
    }

    arrow_table = pa.Table.from_pydict(data)
    
    # Write the data to the Iceberg table
    file_io = PyArrowFileIO()
    with table.new_append(file_io) as writer:
        writer.add(arrow_table)
        writer.commit()
```

```python
if __name__ == "__main__":
    #Check if the consumer is correctly created and not None
    if consumer is not None:
        json_responses = consume_messages(consumer)
        print(json_responses)
        if json_responses:
            table = setup_iceberg_table()
            write_to_iceberg(table, json_responses)
            for json_response in json_responses:
                print(json.dumps(json_response, indent=2))  # Pretty-print JSON response
        else:
            logger.info("No messages consumed.")
else:
    logger.error("Consumer initialization failed. Exiting...")
```
